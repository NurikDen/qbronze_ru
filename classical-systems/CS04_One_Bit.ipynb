В повседневной жизни мы используем десятичную систему счисления. Ее также называют системой с основанием 10, потому что у нас 10 цифр:

$ 0,~1,~2,~3,~4,~5,~6,~7,~8, \mbox{ и } 9  $.

С другой стороны, в информатике широко используется двоичная система счисления, состоящая всего из двух цифр:

$ 0 $ и $ 1 $.

Бит (или двоичная цифра) является основной единицей информации, используемой в информатике и теории информации. 

Его также можно рассматривать как наименьшую "полезную" единицу памяти, которая имеет два состояния, называемые 0 и 1. 

В любой момент бит может находиться либо в состоянии 0, либо в состоянии 1.

<h3> Четыре оператора </h3>

Сколько различных операторов может быть определено для одного бита?

<i>Оператор, в зависимости от текущего состояния бита, обновляет состояние бита (результатом может быть то же самое состояние).</i> 

Мы можем применить четыре различных оператора к одному биту:
<ol>
    <li> Identity (Тождественность): $ I(0) = 0 $ и $ I(1) = 1 $ </li>
    <li> Negation (Отрицание): $ NOT(0) = 1 $ и $ NOT(1) = 0 $ </li>
    <li> Constant (Ноль): $ ZERO(0) = 0 $ и $ ZERO(1) = 0 $ </li>
    <li> Constant (Единица): $ ONE(0) = 1 $ и $ ONE(1) = 1 $ </li>
</ol>
Первый оператор называется оператором Тождественности, так как он не меняет состояние бита.

Второй оператор -- оператор NOT (или НЕ), данные оператор изменяет состояние бита на противоположное. 

<i>Отметим, что состояния бита 0 и 1 можно заменить булевскими значениями False (Ложь) и True (Истина), соответственно. False - это отрицание True и, наоборот, True - отрицание False.</i>

Третий (соответственно и четвертый) оператор возвращает константное значение 0 (соот., 1), вне зависимости от входного значения.

<h3> Табличное представление </h3>

В наших представлениях направление переходов - сверху вниз влево, т.е. начальные состояния находятся сверху, а конечные - слева:

$
    \begin{array}{c|c}
        & начальные~состояния \\ \hline
        конечные~состояния&\hookleftarrow
    \end{array}
$

Мы можем представить действия каждого оператора в виде таблицы:

$
I = \begin{array}{lc|cc} 
    & & начальные & состояния \\ 
    & & \mathbf{0} & \mathbf{1} \\ \hline 
    конечные & \mathbf{0} & \mbox{goes-to} & \emptyset \\  
    состояния & \mathbf{1} & \emptyset & \mbox{goes-to}  \end{array} ,
$
где 
- заголовок (первая строка) представляет начальные значения, и
- первый столбец представляет конечные значения.

Также переходы можно отобразить числами:
- 1, если есть переход между двумя значениями, и, 
- 0, если нет перехода между двумя значениями.

$
I = \begin{array}{lc|cc} 
    & & начальные & состояния \\ 
    & & \mathbf{0} & \mathbf{1} \\ \hline 
    конечные & \mathbf{0} & 1 & 0 \\  
    состояния & \mathbf{1} & 0 & 1  \end{array}
$

Значение, выделенные <b>жирным</b>, представляют начальные и конечные значения битов. Нежирные значение являются переходами.
<ul>
    <li> Левая-верхняя нежирная 1 представляет переход $ 0 \rightarrow 0 $. </li>
    <li> Нижняя-правая нежирная 1 представляет переход $ 1 \rightarrow 1 $. </li> 
    <li> Правый-верхний нежирный 0 означает отсутствие перехода из 1 к 0. </li>
    <li> Нижний-левый нежирный 0 означает отсутствие перехода из 0 к 1. </li>
</ul>
Значения 0 и 1 представляют переходы как True (On) и False (Off), соответственно. 

Аналогично, можно представить и остальные операторы:

$
NOT = \begin{array}{lc|cc} & & начальные & состояния \\ & & \mathbf{0} & \mathbf{1} \\ \hline конечные & \mathbf{0} & 0 & 1 \\  
    состояния & \mathbf{1} & 1 & 0  \end{array}
~~~
ZERO = \begin{array}{lc|cc} & & начальные & состояния \\ & & \mathbf{0} & \mathbf{1} \\ \hline конечные & \mathbf{0} & 1 & 1 \\  
    состояния & \mathbf{1} & 0 & 0  \end{array}
~~~
ONE = \begin{array}{lc|cc} & & начальные & состояния \\ & & \mathbf{0} & \mathbf{1} \\ \hline конечные & \mathbf{0} & 0 & 0 \\  
    состояния & \mathbf{1} & 1 & 1  \end{array}
.
$

<h3> Обратимость (Reversibility) и Необратимость (Irreversibility) </h3>

После применения оператора Тождественности (Identity) или Отрицания (NOT), мы можем определить начальное состояние бита по текущему состоянию. 
<ul>
    <li> В случае оператора Тождественности наш вариант - то же самое значение. </li>
    <li> В случае оператора Тождественности наш вариант - противоположное значение, т.е., Если получен 0 (соот., 1), тогда наш ответ 1 (соот., 0). </li>
</ul>

Однако, мы не можем определить начальное значение по полученному значению после применения оператора ZERO или ONE. 

Основываясь на этом наблюдении, мы можем разделить операторы на два типа: <i>Обратимые</i> и <i>Необратимые</i>.
<ul>
    <li> Если начальные состояния могут быть восстановлены исходя из конечных состояний, то такие операторы называются обратимыми (операторы Identity и NOT). </li>
    <li> Если начальные состояния не могут быть восстановлены исходя из конечных состояний, то такие операторы называются необратимыми (операторы ZERO и ONE). </li>
</ul>

<b> Данная классификация является важной, так как все операторы квантовой эволюции являются обратимыми.</b> (пока система является закрытой).

Операторы Тождественности (Identity) и Отрицания (NOT) - два фундаменталных квантовых оператора.
